### One Sample Poisson Test

This test is used, when we wish to compare the rate at which event occur against a claim value established weather that difference is statistically significant or not.

EX: number of theft happend in local train,

number of chain pull

```python
# A good class would ask atleast 20 question every day
# Ho n >= 20  Ha n < 20  ==> Left tail test

question = [14,19,24,18,17,23,21,18,19,17]

np.mean(question)

test_poisson(sum(question),len(question),20,method='score',alternative='smaller')
```

### Two Sample Poisson Test

This test is used when we have two sample and difference of the average count has to be compare with the claim value.

Ex: number of people buying in AVT(male and female)

```python
'''
DBDA batch ask atleast 6 question/day more than AI batch

Ho avg_count(DBDA) >= 6
Ha avg_count(DBDA) < 6  ==> Left tail test

In DBDA's 12 days of training, 145 question were asked
In AI's 12 days of training, 45 question were asked
'''

test_poisson_2indep(145,12,45,5,value=6,compare='diff',method='score',alternative='smaller')
```

### One Variance Test
Used when the variance of a sample has to be compared against a claim value and establish whether their difference from 1 is statistically significant or not




```python
df = pd.read_excel('CDAC_DataBook.xlsx', sheet_name='faithful')

erupt = df.eruptions
```

```python
np.var(erupt,ddof=1)
```

```python
# Ho variance (erupt) <= 1.1
# Ha variance (erupt) > 1.1    ===> right tail test
# TS = (n-1)*variance/(claim_value)

ts = (len(erupt)-1)*np.var(erupt,ddof=1)/1.1
ts
```

```python

p_value = 1 - chi2.cdf(ts,len(erupt)-1)
p_value
```

### Two Variance Test
Used when we have two samples. The ratio of the variances has to be compared against a claim value and establish whether their difference from 1 is statistically significant or not.


```python
samp1 = erupt[30:70]
samp2 = erupt[130:180]

np.var(samp1,ddof=1)/np.var(samp2,ddof=1)
```

```python
# Ho var(samp1)/var(samp2) >= 1.4
# Ha var(samp1)/var(samp2) < 1.4

samp_ratio = np.var(samp1,ddof=1)/np.var(samp2,ddof=1)
ts = samp_ratio/1.4

f.cdf(ts,len(samp1)-1,len(samp2)-1)
# test-statistic, ddof(smap1), ddof(samp2)  ==> input to f.cdf
```
