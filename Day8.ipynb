Skewness Formula:

sum((x1-xbar)\**3/(n-1)*sd**3)

---
formula:

\sum \frac{(x_i - \bar{x})^3}{(n - 1)s^3}

Kurtosis:

sum((xi-xbar)\**4)/(n*sd**4) -3

Inferential Stats

Sample statistic (SS)
Population Parameter (PP)

- SS = PP (Expected) but not equal

SS = PP + Bias + chance variance

Sample Statistics changes every time we draw a sample but PP remain same

Central limit Therom

- If multiple sample are collected from population and are ploted on a graph then irrespective of population distribution the distribution of the sample mean's will always follow normal distribution.

- Mean of the sample mean is always the population mean.

- Standard deviantion of sample mean would be equal to population std / root(sample)
