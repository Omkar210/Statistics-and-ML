Eigen Vector:
- Direction of the principal components.

Eigen Value:
- Amount of variance explained in the direction of eigen vectors.

Co-Variance:
- magnitude of co-variance is useless, it can be (-inf,inf)
- sign is important
- it tells you about the +ve or -ve corelation between two factors but nothe magnitude.
- Formula:
$$
Co-Variance (COV) = \frac{\Sigma^{N}_{i=1} (X_i - \mu_x)(Y_i - \mu_y)}{n - 1}
$$
$$
Co-Relation Coeffient(CORR) = \frac{COV}{\sigma_1 * \sigma_2}
$$

---

Characteristic Equation:
$$
|R - \lambda I| = 0
$$
R => Corealtion Matrix
I => Identity Matrix
$\lambda$ => eigen value
- Standardization of the values are necessary for PCA and LDA
- power of lambda will be decided on number of features.


# Principal Component Analysis (PCA) - Manual Calculation Using Characteristic Equation

## Step 1: Given Dataset

| Feature 1 | Feature 2 | Feature 3 |
|------------|------------|------------|
| 2 | 6 | 8 |
| 4 | 4 | 5 |
| 5 | 3 | 9 |
| 2 | 3 | 9 |
| 7 | 8 | 7 |

Let

$$
X=
\begin{bmatrix}
2 & 6 & 8\\
4 & 4 & 5\\
5 & 3 & 9\\
2 & 3 & 9\\
7 & 8 & 7
\end{bmatrix}
$$

Number of observations:

$$
n=5
$$

Number of features:

$$
p=3
$$

---

# Step 2: Compute Mean of Each Feature

For Feature 1:

$$
\bar{x}_1=\frac{2+4+5+2+7}{5}
=\frac{20}{5}
=4
$$

For Feature 2:

$$
\bar{x}_2=\frac{6+4+3+3+8}{5}
=\frac{24}{5}
=4.8
$$

For Feature 3:

$$
\bar{x}_3=\frac{8+5+9+9+7}{5}
=\frac{38}{5}
=7.6
$$

Mean Vector:

$$
\mu=
\begin{bmatrix}
4\\
4.8\\
7.6
\end{bmatrix}
$$

---

# Step 3: Mean Center the Dataset

Subtract column means from each feature.

$$
X_{centered}
=
X-\mu
$$

$$
=
\begin{bmatrix}
-2 & 1.2 & 0.4\\
0 & -0.8 & -2.6\\
1 & -1.8 & 1.4\\

3 & 3.2 & -0.6

---

# Step 4: Compute Covariance Matrix

Formula:

$$
C=
\frac{1}{n-1}
X_{centered}^{T}
X_{centered}
$$

Since:

$$
n=5
$$

$$
n-1=4
$$

Computing:

$$
X_{centered}^{T}
X_{centered}
=
\begin{bmatrix}
18 & 11 & -2\\
11 & 18.8 & -6.4\\
-2 & -6.4 & 12.8
\end{bmatrix}
$$

Therefore:

$$
C=
\frac{1}{4}
\begin{bmatrix}
18 & 11 & -2\\
11 & 18.8 & -6.4\\
-2 & -6.4 & 12.8
\end{bmatrix}
$$

Covariance Matrix:

$$
C=
\begin{bmatrix}
4.50 & 2.75 & -0.50\\
2.75 & 4.70 & -1.60\\
-0.50 & -1.60 & 3.20
\end{bmatrix}
$$

---

# Step 5: Find Eigenvalues Using Characteristic Equation

PCA requires solving:

$$
|C-\lambda I|=0
$$

Substitute covariance matrix:

$$
\left|
\begin{matrix}
4.5-\lambda & 2.75 & -0.5\\
2.75 & 4.7-\lambda & -1.6\\
-0.5 & -1.6 & 3.2-\lambda
\end{matrix}
\right|
=0
$$

Expanding determinant:

$$
(4.5-\lambda)
\left[
(4.7-\lambda)(3.2-\lambda)
-
(-1.6)^2
\right]
$$

$$
-
2.75
\left[
2.75(3.2-\lambda)
-
(-1.6)(-0.5)
\right]
$$

$$
+
(-0.5)
\left[
2.75(-1.6)
-
(4.7-\lambda)(-0.5)
\right]
=0
$$

After simplification:

$$
\lambda^3
-
12.4\lambda^2
+
41.1575\lambda
-
25.242
=
0
$$

Characteristic Equation:

$$
\boxed{
\lambda^3
-
12.4\lambda^2
+
41.1575\lambda
-
25.242
=
0
}
$$
