## Interval Estimate for population

An interval estimate (or confidence interval) provides a range of values within which a population parameter (like the mean or proportion) is likely to fall. It is calculated using sample data, giving you a statistical measure of how accurate or reliable your estimate is.

```python
# this is how python does for calculating range of interval
n = 300
xbar = 1800
stdev = 70
cl = 0.95

norm.ppf((1-cl)/2)
```

```python
z = np.abs(norm.ppf((1-cl)/2))
z  # critical value of z
```

```python
# t distribution generally not required but python uses so it's needed
t1 = np.abs(t.ppf((1-cl)/2,n-1))
t1  # critical value of t1
```

```python
low_est = xbar - t1*stdev/n**0.5
upper_est = xbar + t1*stdev/n**0.5

size = upper_est - low_est
size
```

```python
# this is how python does for calculating range of interval
n = 3000
xbar = 1800
stdev = 70
cl = 0.95

norm.ppf((1-cl)/2)

z = np.abs(norm.ppf((1-cl)/2))
z  # critical value of z

t1 = np.abs(t.ppf((1-cl)/2,n-1))
t1  # critical value of t1

low_est = xbar - t1*stdev/n**0.5
upper_est = xbar + t1*stdev/n**0.5

size = upper_est - low_est
size
```

```python
# direct function for calculating range 0f interval
stats.t.interval(cl,n-1,xbar,stdev/n**0.5)
        # confidence level, sample size(ddof), xbar, std error of mean
```
