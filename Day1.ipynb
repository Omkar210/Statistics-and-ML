```python
# Broad-casting: make an array different from list
b1 = np.array([4,7,6,8,9])
b2 = np.array([4,2,7,6,3])

b1 + 5
# Broadcasting make it possible to repeat the same operation across all the values

2*b1 + 5
b1 + b2
```
