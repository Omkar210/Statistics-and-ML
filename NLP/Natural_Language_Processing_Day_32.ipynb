# BERT - Bidrectional Encoder Representation from Transformer

## Definition

**BERT (Bidirectional Encoder Representations from Transformers)** is a pre-trained language model developed by Google in 2018. It uses **only the Transformer Encoder** and reads text **bidirectionally** (from both left and right simultaneously) to understand the context of each word.

---

# Why was BERT introduced?

Older models like RNN, LSTM, and even early Transformer models had limitations:

* RNN/LSTM struggled with long-range dependencies.
* Traditional language models often read text in only one direction (left-to-right or right-to-left).
* The meaning of a word depends on both its previous and following words.

BERT solves this by reading the **entire sentence at once**.

---

# Example

Sentence:

```text
The bank is near the river.
```

Here, **bank** means the side of a river.

Now another sentence:

```text
I deposited money in the bank.
```

Here, **bank** means a financial institution.

BERT understands the correct meaning because it considers **both the left and right context** of the word.

---

# What does "Bidirectional" mean?

Traditional language model:

```text
I love AI

Predict "AI"

Reads:

I → love → AI
```

Only previous words are used.

---

BERT:

```text
I love AI

      AI

Left context  : I love
Right context : (sentence end)

Uses BOTH directions.
```

For every word, BERT considers all surrounding words.

---

# Architecture

BERT uses **only the Encoder** part of the Transformer.

```text
Input Sentence

↓

Embedding Layer

↓

Transformer Encoder
Layer 1

↓

Transformer Encoder
Layer 2

↓

...

↓

Transformer Encoder
Layer N

↓

Contextual Word Representations
```

Unlike encoder-decoder models, **BERT has no decoder**.

---

# Input Representation

Each input token is represented by adding three embeddings:

```text
Token Embedding
        +
Position Embedding
        +
Segment Embedding
```

---

### 1. Token Embedding

Converts each word into a vector.

Example:

```text
I

↓

[0.2, 0.5, 0.1, ...]
```

---

### 2. Position Embedding

Transformers do not process words sequentially, so position information is added.

Example:

```text
Word      Position

I             1

love          2

AI            3
```

---

### 3. Segment Embedding

Used when there are two sentences.

Example:

```text
Sentence A

What is AI?

Sentence B

Artificial Intelligence
```

Sentence A gets one segment ID, and Sentence B gets another.

---

# Special Tokens

### `[CLS]`

Placed at the beginning.

```text
[CLS] I love AI
```

The final representation of `[CLS]` is used for **sentence-level tasks**, such as classification.

---

### `[SEP]`

Separates sentences.

Example:

```text
[CLS]

What is AI?

[SEP]

Artificial Intelligence

[SEP]
```

---

### `[MASK]`

Used during training.

Example:

```text
I love [MASK].
```

The model predicts the missing word.

---

# Pre-training Tasks

## 1. Masked Language Modeling (MLM)

Some words are hidden.

Example:

```text
I love [MASK].
```

Prediction:

```text
AI
```

Since the model sees words on both sides of `[MASK]`, it learns bidirectional context.

---

## 2. Next Sentence Prediction (NSP)

BERT learns whether one sentence logically follows another.

Example:

Sentence A

```text
I went to school.
```

Sentence B

```text
I attended my classes.
```

Prediction:

```text
Is Next Sentence?

YES
```

Another example:

Sentence A

```text
I bought a laptop.
```

Sentence B

```text
The moon is made of cheese.
```

Prediction:

```text
NO
```

---

# Attention in BERT

Each word attends to every other word.

Example:

```text
The cat sat on the mat.
```

When processing:

```text
cat
```

The model can pay attention to:

```text
The
cat
sat
on
the
mat
```

This helps capture relationships regardless of distance.

---

# Training Phase

During pre-training:

```text
Input

I love [MASK]

↓

Encoder

↓

Predict

AI
```

Millions of sentences are used so BERT learns grammar, syntax, and semantics.

---

# Fine-Tuning

After pre-training, BERT can be adapted for many tasks by adding a small task-specific output layer.

```text
Pre-trained BERT

↓

Fine-Tuning

↓

Specific Task
```

Examples:

* Sentiment Analysis
* Question Answering
* Named Entity Recognition (NER)
* Text Classification
* Spam Detection

---

# Advantages

* Understands context from both directions.
* Produces contextual word embeddings.
* Achieves high accuracy on many NLP tasks.
* Handles long-range dependencies using self-attention.
* Can be fine-tuned for many applications.

---

# Limitations

* Computationally expensive.
* Requires significant memory.
* Slower than smaller language models.
* Primarily designed for understanding text rather than generating long passages.

---

# Applications

* Sentiment Analysis
* Question Answering
* Machine Translation (mainly as an encoder)
* Named Entity Recognition
* Text Classification
* Search Engines
* Chatbots
* Document Classification

---

# BERT vs GPT

| Feature               | BERT                            | GPT                                      |
| --------------------- | ------------------------------- | ---------------------------------------- |
| Transformer Part Used | Encoder only                    | Decoder only                             |
| Reading Direction     | Bidirectional                   | Left-to-right (autoregressive)           |
| Main Purpose          | Language understanding          | Text generation                          |
| Training Objective    | Masked Language Modeling (MLM)  | Next-token prediction                    |
| Generates text        | Limited                         | Excellent                                |
| Best for              | Classification, NER, QA, search | Chatbots, story writing, code generation |

---

## Key Points to Remember

* **BERT = Bidirectional Encoder Representations from Transformers.**
* Uses **only the Transformer Encoder**.
* Reads text **in both directions simultaneously**.
* Trained using **Masked Language Modeling (MLM)** and **Next Sentence Prediction (NSP)**.
* Uses special tokens: **`[CLS]`**, **`[SEP]`**, and **`[MASK]`**.
* Produces **contextual embeddings**, so the same word can have different meanings depending on the sentence.
* Best suited for **language understanding tasks** rather than autoregressive text generation.

````python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import tensorflow as tf

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(
    "distilbert-base-cased-distilled-squad"
)

model = AutoModelForQuestionAnswering.from_pretrained(
    "distilbert-base-cased-distilled-squad"
)

question = "What is my name?"
context = "My name is Ruchika and I work at GA Services in Bangalore."

# Tokenize
inputs = tokenizer(
    question,
    context,
    return_tensors='pt'
)

print(inputs)

# Forward pass
outputs = model(**inputs)
print(outputs)

# Use PyTorch's argmax and .item() for PyTorch tensors
start = outputs.start_logits.argmax(dim=1).item()
end = outputs.end_logits.argmax(dim=1).item() + 1

answer = tokenizer.decode(
    inputs["input_ids"][0][start:end],
    skip_special_tokens=True
)

print("Question:", question)
print("Context:", context)
print("Answer:", answer)
````

````python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Load tokenizer and model explicitly with from_pt=True
tokenizer = AutoTokenizer.from_pretrained("sshleifer/distilbart-cnn-12-6")
model = AutoModelForSeq2SeqLM.from_pretrained("sshleifer/distilbart-cnn-12-6", from_pt=True)

summarizer = pipeline(
    "summarization",
    model=model, # Pass the loaded TF model
    tokenizer=tokenizer # Pass the loaded tokenizer
)

text = """
*Dilwale Dulhania Le Jayenge (DDLJ)* is a landmark Bollywood romantic film released in 1995.
     Directed by Aditya Chopra and starring Shah Rukh Khan and Kajol, the movie tells the story of Raj and Simran,
      two young Indians who meet and fall in love during a trip across Europe. Their journey is filled with romance, humor,
       and emotional moments that have captivated audiences for decades.

The film beautifully blends love, family values, and tradition, emphasizing the importance of respecting one's parents
while pursuing personal happiness. Known for its unforgettable music, memorable dialogues, and strong performances,
 DDLJ became one of the biggest successes in Indian cinema and continues to be regarded as one of the greatest Bollywood films
  ever made.


"""

summary = summarizer(
    text,
    max_length=150,
    min_length=30,
    do_sample=False
)

print(summary[0]["summary_text"])
````
