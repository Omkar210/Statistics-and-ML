```python
from sklearn.feature_extraction.text import CountVectorizer
# Sample documents
documents = [
    "I love NLP",
    "NLP is interesting",
    "I love machine learning"
]

# Create BoW model
vectorizer = CountVectorizer() #creates vocabulary


# Convert text into BoW matrix
X = vectorizer.fit_transform(documents)#counts occurances based on vocabulary

# Display as table
df = pd.DataFrame(
    X.toarray(),#sparse matrix to normal
    columns=vectorizer.get_feature_names_out()
)

print(df)
```

TF:How often a word appears in a document \
IDF:Gives less importance to words appearing in many documents.

```python
from sklearn.feature_extraction.text import TfidfVectorizer


# Sample documents
documents = [
    "I love NLP",
    "NLP is interesting",
    "I love machine learning"
]

# Create TF-IDF model
vectorizer = TfidfVectorizer()

# Convert text into TF-IDF matrix
X = vectorizer.fit_transform(documents)

# Display as table
df = pd.DataFrame(
    X.toarray(),
    columns=vectorizer.get_feature_names_out()
)

print(df.round(2))
```

For Document 0, the most important terms are "love" and "nlp" (TF-IDF = 0.71).

For Document 1, the most important terms are "interesting" and "is" (TF-IDF = 0.62).

For Document 2, the most important terms are "machine" and "learning" (TF-IDF = 0.62).

Words with higher TF-IDF values contribute more to distinguishing one document from the others.

So TF-IDF is often preferred over Bag of Words for text classification and information retrieval.

```python
import pandas as pd
import nltk

import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag

from nltk.corpus import stopwords
import re
from collections import Counter


nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report
```

```python
df=pd.read_csv("/content/drive/MyDrive/CDAC/spam.csv",encoding="latin1")
```

```python
df.head()
```

```python
df=df.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"],axis=1)
```

```python
df.head()
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(text):

    # lowercase
    text = text.lower()

    # remove special characters except englih characters
    text = re.sub(r'[^a-zA-Z]', ' ', text)

    # tokenize
    words = nltk.word_tokenize(text)

    # remove stopwords and lemmatize
    words = [
        lemmatizer.lemmatize(word)
        for word in words
        if word not in stop_words
    ]

    return " ".join(words)
```

```python
df.iloc[4]
```

```python
df['clean']=df['message'].apply(preprocess)
```

```python
df[25:50]
```

```python
df['label'] = df['label'].map({
    'ham': 0,
    'spam': 1
})
```

```python
df['label'].value_counts()
```

```python
X = df['message']
y = df['label']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced',max_iter=1000)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
print("Accuracy:", accuracy_score(y_test, predictions))

print(classification_report(y_test, predictions))
```

```python
new_emails = [

     "Hi Chetan, the meeting is scheduled at 10 AM tomorrow.",

    "Claim your prize money immediately."
]
```

```python
clean_emails = []

for email in new_emails:
    clean_emails.append(preprocess(email))
```

```python
email_vectors = tfidf.transform(clean_emails)
```

```python
predictions = model.predict(email_vectors)
```
