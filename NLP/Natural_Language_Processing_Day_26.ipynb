```python
import nltk
from nltk.stem import PorterStemmer
stem1=PorterStemmer()
words=['run','running','ran','runs','easily','fairly','happy','affection','caring']
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lem1 = WordNetLemmatizer()
```

```python
for i in words:
    print(i+' : '+stem1.stem(i)+' Stem')
    print(i+' : '+lem1.lemmatize(i)+' Lemma')
```

```python
# Part of Speech

lem1.lemmatize('better', pos='a')  # Adjective
```
