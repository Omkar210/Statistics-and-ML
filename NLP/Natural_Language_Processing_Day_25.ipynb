````python
import re
````

````python
text = 'NLP is amazing! Visit us at https://npl.ai or call +-800-555-0199 today'
````

````python
text = text.lower()
print(text)
````

````python
text = re.sub('https?://\S+','',text)
print(text)
````

````python
text = re.sub(r'\+?\d[\d\s\-]{6,}','',text)
print(text)
````

````python
text = re.sub(r'[!"#$%&()*+,-./:;<=>?@[\]^_`{|}~]', '', text)
print(text)
````

````python
text = re.sub(r'[^a-z\s]','',text)
print(text)
````

````python
text = re.sub(r'\s+','',text)
print(text)
````
