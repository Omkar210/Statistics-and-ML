# Feature Scaling

Techniques:
- Standardization:  original value - mean / sd, Profit: put the values around 0.

- Standardization formula:  x - μ ÷ σ

- Min-Max Scaler: (x - min(col)) / (max(col) - min(value))

- Robust Scaler : (x - median(col)) / (percentile(col,75) - percentile(col,25))

standardize columns

It uses the population sd rather than sample sd which will be little fluctuation

MinMaxScaler

If there is outlier then it may differ

```python
df = pd.DataFrame({'Values':[3,6,5,1,6,8,3,8,9,4]})

df['StdScale'] = (df.Values-np.mean(df.Values))/np.std(df.Values,ddof=1)

df['MinMaxScale'] = (df.Values-np.min(df.Values))/(np.max(df.Values)-np.min(df.Values))

df['RobustScale'] = (df.Values-np.median(df.Values))/(np.percentile(df.Values,75)-np.percentile(df.Values,25))

df
```

### **Categorical Regression:**
Response continuous, predictors atleast one categorical and other can be continuous.

* Response Continuous, Predictors will be categorical and continuous.

One Hot Encode:

smaller alphabet will become the reference event.


```python
df = pd.read_excel('CDAC_DataBook.xlsx', sheet_name = 'salaries')

df = df[['yrs_service','gender','salary']]

gender_dum = pd.get_dummies(df.gender,drop_first=True).astype(int)

df = pd.concat([df,gender_dum], axis=1)

df = df.drop('gender', axis=1)


# Split for train-test
x_train, x_test, y_train, y_test = train_test_split(df.drop('salary', axis=1),df.salary, test_size=0.2, random_state=20)

# Add constant term
x_train = sm.add_constant(x_train, prepend=False)
x_test = sm.add_constant(x_test, prepend=False)

# Prepare the model
mod1 = sm.OLS(y_train,x_train).fit()

print(mod1.summary())


# Ho yrs_service = 0
# pvalue = 0 reject
# reference value will never appeare in OLS summary
#
# Ho : with respect to female, male does not influence salary
# Ho : With respect to female(reference event), if male is selected then it soes not influence the salary

# With respect to female, if same numbers of experience increase then male's salary decreased by 6000.
```
