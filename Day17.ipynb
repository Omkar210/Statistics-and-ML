```python
df = pd.read_excel('CDAC_DataBook.xlsx', sheet_name = 'diabetes')

df =df[['Glucose','BloodPressure','Age','DietType','Outcome']]

x = df.drop('Outcome', axis=1)
y = df.Outcome

diet_dummy = pd.get_dummies(x.DietType, drop_first=True, prefix='Diet').astype(int)
x = x.drop('DietType', axis=1)
x = pd.concat([x, diet_dummy], axis=1)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=20)

x_train = sm.add_constant(x_train, prepend=False)
mod3 = sm.Logit(y_train, x_train).fit()
print(mod3.summary())
```

Precision :
$$
\text{Precision} = \frac{TP}{TP + FP}
$$

$$
\text{Precision} = \frac{TN}{TN + FN}
$$

Recall :
$$
\text{Recall} = \frac{TP}{TP + FN}
$$

$$
\text{Recall} = \frac{TN}{TN + FP}
$$

F1 - Score :
$$
F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}
{\text{Precision} + \text{Recall}}
$$

```python
x_test = sm.add_constant(x_test, prepend=False)
y_pred_prob = mod3.predict(x_test)
y_pred = []

for i in y_pred_prob:
  if i >= 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)

print('''True Negative     False Positive
False Negative    True Positive\n''')

print(f'Confusion Matrix: \n{confusion_matrix(y_test, y_pred)}\n\n')
print(f'Classification Report: \n{classification_report(y_test, y_pred)}')
```

# One Hot Encoding

- Is used for Nominal scale variable - no logical order
- Not useful when numner of categories are too large
- Regression, ANN, KNN, SVM

# Ordinal Encoding

- Variable with inheritance rankinh or rating
- preseve logical order


```python
df = pd.DataFrame({
    'Education':[
        'Graduate',
        'PG',
        'High School',
        'Graduate',
        'PG'
    ]
})

education_map = {
    'High School':1,
    'Graduate':2,
    'PG':3
}

df['Education_Encoded'] = (
    df['Education']
    .map(education_map)
)

print(df)
```

```python
# Frequency encoding
df = pd.DataFrame({
    'City':[
        'Mumbai',
        'Mumbai',
        'Mumbai',
        'Delhi',
        'Delhi',
        'Pune'
    ]
})

freq = df.City.value_counts()

df['City_Frequency'] = (df.City.map(freq))
df
```

```python
df = pd.DataFrame({
    'City': [
        'Mumbai',
        'Pune',
        'Delhi',
        'Chennai',
        'Kolkata'
    ]
})

encoder = BinaryEncoder(cols=['City'])
df_encoded = encoder.fit_transform(df)
print(df_encoded)

df_encoded = encoder.transform(df)
result = pd.concat([df, df_encoded], axis=1)
print(result)
```
