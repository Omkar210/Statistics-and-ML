# Tests on Categorical

### One sample Z-Test or One Sample T-Test

Z-test: only for continuous data, sample mean is > 30.

- Used to check whether the difference between the sample mean and claim value can be ignored or not.

- Used to check whether the difference between the sample mean and claim value is statistically significant or not

Step 1:
Hₒ : μ >= 3.7
Hₐ : μ < 3.7

Step 2:
CL = 95% = 0.95
CL = 5% = 0.05

Step 3:
Because Mean has to test with Claim mean.

Step 4:
Test Statistics , p-value

Step 5:


```python
ssw.ztest(erupt, value=3.7, alternative='smaller') # calculating the z-value and p-value respectively
```

OR

```python
z = (xbar - 3.7)/(sd/n**0.5)
z
```

```python
# 5% value on x-axis is called Zc
norm.ppf(0.05)
```

start point of 5% on standard normal curve is -1.6448536269514729

and z is -3.066461475706521 and p-value(0.0010830435959524984)

because the z-value is lesser than start point of 5% so we reject the NULL HYPOTHESIS.

Upto -1.645 is okay to be DO NOT REJECT.

---

There is 2 options to choose between Rejection or not to NULL HYPOTHESIS
1. Compare with z-value:

**Rejection will happen if z-value is more than 5%**

z-value is value on x-axis, if it's magnitude is more than Zc then we REJECT

2. Compare with p-value:

p-value is percentage, if it's magnitude is less than LOS then REJECT.

---
Calculation:

p-value -> norm.cdf(z)

```python
# Claim mean is 3.53

p = ssw.ztest(erupt, value=3.53, alternative='smaller')

# OR

z = (xbar - 3.53)/(sd/n**0.5)
print(f'z-value = {z}, \nz-value(stats.model) = {p[0]}, \np-value(stats.model) = {p[1]}')
```

```python
# p-value
norm.cdf(z)
```

```python
# Null Hypthesis : mean < 3.37

print(ssw.ztest(erupt, value=3.37, alternative='larger')) # larger becz right tail test

print((xbar-3.37)/(sd/n**0.5))  # z-value

print(norm.ppf(0.95))  # ppf will take the left side area

print(1-norm.cdf(1.7019251649638607))  # cdf will take the z-value for calculating the p-value.
```

```python
'''
Q. min waiting time 75min
test stat?
p-value?
95% ub
'''

wait = df.waiting
xbar = np.mean(wait)
n = len(wait)
sd = np.std(wait,ddof=1)

z = (xbar-75)/(sd/n**0.5)

print(z) # test statistics
print(norm.cdf(z)) # p-value
print(norm.ppf(0.95))  # area of 95%
print(xbar + norm.ppf(0.95)*sd/n**0.5)  # 95% Upperbound

# OR

print(ssw.ztest(wait, value=75, alternative='smaller'))  # left tail test so 'smaller'
```

# 2 sample t-test

this test is used when the difference of means of two independent samples has to be compared against the claim value and established whether their difference is statistically significant or not.

Area of Usage:

Two Independant Sample

```python
'''
Travel in Blore is atleast 15 min more than Mumbai

Blore
Sample size 50
Avg time 45
Std Dev 10.5

Mumbai
Sample size 60
Avg time 34
Std Dev 5.6
'''
```

```python
df1 = df.iloc[20:60]
df2 = df.iloc[120:170]

df1.shape
```

## mean of erupt 2 is atleast 0.12 more than the mean of erupt1

Hₒ : mean(erupt2) - mean(erupt1) >= 0.12

Ha : mean(erupt2) - mean(erupt1) < 0.12

```python
erupt1 = df1.eruptions
erupt2 = df2.eruptions
```

```python
ssw.ztest(erupt2,erupt1,value=0.12,alternative='smaller')
```

```python
xbar = np.mean(erupt1)
n1 = len(erupt1)
sd1 = np.std(erupt1,ddof = 1)
n2 = len(erupt2)
sd2 = np.std(erupt2,ddof = 1)
Nr = ((np.mean(erupt2) - np.mean(erupt1)) - 0.12)
Dr = ((sd1**2/n1) + (sd2**2/n2))**0.5

z = Nr/Dr
z
```
