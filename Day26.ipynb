# K-Nearest Neighbors (KNN) Classifier

## Key Points

- KNN is a **distance-based supervised learning algorithm**.
- It classifies a new data point based on the majority class of its nearest neighbors.
- **Feature scaling is mandatory** because distance calculations are sensitive to feature magnitudes.
- Common scaling techniques:
  - Standardization (Z-score Normalization)
  - Min-Max Scaling

## Euclidean Distance

The Euclidean distance between two points is:

$$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

For \(n\)-dimensional data:

$$
d = \sqrt{\sum_{i=1}^{n}(a_i - b_i)^2}
$$

## Manhattan Distance

The Manhattan distance between two points is:

$$
D = \sum_{i=1}^{n} |a_i - b_i|
$$

## KNN Classification Process

1. Choose the value of \(K\).
2. Calculate the distance between the new data point and all training points.
3. Select the \(K\) nearest neighbors.
4. Count the class labels of those neighbors.
5. Assign the class with the majority vote.

## Example

Suppose:

| Point | Class |
|--------|--------|
| (2,3) | A |
| (3,4) | A |
| (8,7) | B |
| (9,8) | B |

New Point:

$$
P = (4,4)
$$

Using Euclidean distance:

$$
d((4,4),(2,3))
=
\sqrt{(4-2)^2+(4-3)^2}
=
\sqrt{5}
$$

$$
d((4,4),(3,4))
=
\sqrt{(4-3)^2+(4-4)^2}
=
1
$$

The nearest neighbors belong to Class **A**, so KNN predicts:

$$
\text{Class}(P) = A
$$

## Advantages

- Simple and easy to understand.
- No training phase (lazy learner).
- Works well for small datasets.
- Can handle multi-class classification.

## Disadvantages

- Computationally expensive for large datasets.
- Sensitive to irrelevant features.
- Requires feature scaling.
- Performance depends heavily on the choice of \(K\).

```python
df = pd.read_excel('CDAC_DataBook.xlsx', sheet_name = 'diabetes')
df = df[['Glucose', 'BloodPressure', 'Age', 'DietType', 'Outcome']]
x = df.drop('Outcome', axis=1)
y = df.Outcome

diet_dummy = pd.get_dummies(df['DietType'], drop_first=True, prefix='DT')
x.drop('DietType', axis=1)
x = pd.concat([x, diet_dummy], axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=20)

mod1 = KNeighborsClassifier(n_neighbors=7)
mod1.fit(x_train, y_train)
y_pred = mod1.predict(x_test)
print(confusion_matrix(y_test, y_pred))
```
