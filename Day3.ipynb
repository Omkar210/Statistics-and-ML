```python
df = pd.read_excel('CDAC_DataBook.xlsx',sheet_name='EmpInfo')
df.head()
```

Department Column have null Values or not

```python
ind = np.where(df.Deptt.isnull())
ind
```

Create two different dataframe where Department is missing or not respectively

```python
df1 = df.iloc[np.where(df.Deptt.isnull())[0]].reset_index(drop=True)
df1
```

```python
df2 = df.drop(np.where(df.Deptt.isnull())[0]).reset_index(drop=True)
df2
```

Deptt and EmpId is missing

```python
df.iloc[np.where(df.Deptt.isnull() & df.EmpID.isnull())[0]].reset_index(drop=True)
```

Change where Deptt is null value to Fresher

```python
df.iloc[np.where(df.Deptt.isnull())[0],2] = 'Fresher'
df
```

Don't want any missing values in the dataset then use dropna()

```python
temp = df.dropna()
temp
```
