Where Quantity is greater than 70

```python
print(len(df.iloc[np.where(df.Quantity>70)]))
print(df.iloc[np.where(df.Quantity>70)])
print(df.iloc[np.where(df.Quantity>70)].count())
print((df.Quantity>70).sum())

ind = np.where(df.Quantity>70)[0]
df.iloc[ind]
```

```python
ind = np.where((df.Quantity>=70) & (df.Location=='MWH-1'))
print(len(ind[0]))

df2 = df.iloc[ind]
print(df2)
```

```python
ind = np.where((df.Quantity>=70) | (df.Location=='MWH-1'))
print(len(ind[0]))

df3 = df.iloc[ind]
print(df3.head())
```

```python
# Weight of location MWH-2
n2 = df.iloc[np.where(df.Location=='MWH-2')]
print(sum(n2.Quantity))

sum(df.iloc[np.where(df.Location=='MWH-2')].Quantity)
```
