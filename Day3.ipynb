GroupBy Command

```python
grp1 = df.groupby('Location')

# Grouping has been done based on the location
print(grp1.groups)

# print perticular group
df1 = grp1.get_group('MWH-3').reset_index(drop=True)
print(df1)

# print multiple group at a time
print(pd.concat([grp1.get_group('MWH-2'),grp1.get_group('MWH-3')]))

# sum of all Quantity present in each group
# reset_index will get you label in column and index is independent
df2 = grp1.Quantity.agg('sum').reset_index()
print(df2)

# this will give you how many time that record is in dataframe
# counting the total number of records are present in each groups
print(grp1.Quantity.agg('size'))
print(grp1.Quantity.agg('mean'))
print(grp1.Quantity.agg('median'))

# multiple aggregate function is used in one function
print(grp1.Quantity.agg(['sum','mean','size']))
```

Multi-Index or Multi-level labels

```python
locn1 = ['Mumbai','Mumbai','Mumbai','Pune','Pune']
course = ['AI','ML','DS','AI','DS']
marks1 = [45,78,65,35,89]
marks2 = [80,76,59,82,49]

mydf = pd.DataFrame({'Omkar':marks1,'Chaitu':marks2},index=[locn1,course])
mydf
```

```python
# index Name will be applied to labels
# this type label/index is called multi-index where index name is empty by default
mydf.index.names = ['Location','Subject']
mydf
```

```python
# print on basis of primary index
print(mydf.loc['Mumbai'])

# print only for AI Subject from Mumbai region using both level of index/label
mydf.loc[('Mumbai','AI')]
```

```python
# only in case of secondary index, we have to use .xs method
mydf.xs('AI',level='Subject')
```
