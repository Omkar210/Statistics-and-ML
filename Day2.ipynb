```python
import numpy as np
import pandas as pd
```

```python
a1 = np.array([3,7,6,1,8,9])
a2 = np.array([2,4,5,8,10,3,6])

print(np.intersect1d(a1,a2))
print(np.union1d(a1,a2))

print(np.in1d(a1,a2)) # DeprecationWarning: instead use below command
print(np.isin(a1,a2))
```

```python
print(np.setdiff1d(a1,a2))
print(np.setdiff1d(a2,a1))
```

```python
# either only in a1 or in a2
print(np.setxor1d(a1,a2))
print(np.where(np.isin(a1,a2))) # this is intersection of a1 and a2
```

# Pandas

```python
import pandas as pd
```

```python
x1 = [4,8,4,3,2]
s1 = pd.Series(x1)
s1 = pd.Series(x1,index=['A','B','C','D','E'])
s1
```

```python
print(s1.iloc[2]) # index based extraction
print(s1.loc['D']) # label based extraction
print(s1.loc[['A','B']])
print(s1.iloc[[2,3]])
```

```python
s2 = pd.Series([1,2,3,4,5], index=['A','A','B','A','B'])
print(s2.iloc[3])
print(s2.loc['A'])
```

```python
s3 = pd.Series([3,9], index=['F','G'])
s4 = pd.concat([s1,s3])
print(len(s4))
s4
```

```python
s4.iloc[4] = 7
print(s4)

s4.loc['E'] = 11
print(s4)
```

```python
s5 = pd.concat([s2,s3])

s5.iloc[2] = 8
print(s5)

s5.loc['A'] = 13
print(s5)
```

```python
print(s4.drop(['D']))
# s5 = s5.drop(['G'])
print(s4)
print(s5)

s5.drop(s5.index[[2,4]])
```

```python
s4 = pd.Series([1,2,4,5,9,4,2,5,7,3,5,5,4,6,7,7,8])
s5 = s4.value_counts()
print(s5)
```

```python
print(s5.iloc[2])
print(s5.loc[5])

# What if we only have values then how to find
np.where(s5==3)
s5.index[np.where(s5==3)]
```

# DataFrame

```python
mydata = [
    ['A',1234,80000],
    ['B',4567,10000],
    ['C',6789,12888],
    ['D',5678,13000]
]
```

```python
# indices are getting carry forward in this dataframe
mydf = pd.DataFrame(mydata)
mydf = pd.DataFrame(mydata, columns=['Name','AccNo','Balance'])
temp = pd.DataFrame(mydata, columns=['Name','AccNo','Balance'],index=['Omkar','Chaitu','Sujal','Chikki'])

print(mydf)
print(temp)
```

```python
df = mydf.loc[[1,2]]
print(df)
print(df.iloc[1])
print(df.loc[1])
```

```python
# label from the original dataframe will get reset index
df = mydf.loc[[1,2]].reset_index(drop=True) # this will reset index

df = mydf.loc[[1,2]].reset_index(drop=False) # this will reset index but create new column of previous label

print(df)
```
