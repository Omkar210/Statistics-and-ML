# **Type of Data**

### 1. Qualitative / Categorical / Attribute: Mostly Categorical Data(Strings)

### 2. Quantitative: Numbers(Age, Scores)


- Continuous Data: Is the one where possibility of decimal point cannot be ruled out.
ex: will always be the measure and the measurment will be coming from instrument -> Weight, Height

- Discrete Data: Data where possibility of decimal point is completely ruled out.
ex: will always be the result of count and will never have a decimal point

- when n*p and n*(1-p) then discreate data can be considered as continuous

- Any Rating or Ranking is number is categorical data

**Googly Question:** Money related Data will Discrete.

# **Types of Scale**

1. Nominal: What is your Gender? (Male or Female) <- Category
Category don't have logical order (Female or Male) there will be same Answer

2. Ordinal: Order Matters
Rating in 1 to 5, 1 is poor and 5 is excellent and if 1 is Excellent and 5 is poor then the answer will be change.

3. Interval: 0(zero) doesn't mean absence of measure, so mathematical calculation/operation will be restricted. Ex: Temperature: 0 C => 0 is like 0 deg temp but zero mean not like temp is not exist.
- Cannot take Ratio
- They are comparable

4. Ratio

Ratio = p/q both are countinuous ex: dist / time

proportion = p/q both are discrete ex: number of poeple from maharashtra / total number of people in class is discrete.

```python
df = pd.read_excel('CDAC_DataBook.xlsx',sheet_name='ERPData')
df.head()
```

```python
x1 = df.MaterialID
```

```python
np.unique(x1)
```

```python
len(np.unique(x1))
```

```python
np.unique(x1, return_counts=True)
```

```python
x1.value_counts()
```

```python
print(x1.value_counts().reset_index(drop=True))
df1 = x1.value_counts().reset_index()
df1
```

```python
df1['prop'] = df1['count']/sum(df1['count'])
df1
```

```python
df1['pct'] = df1['prop']*100
df1
```

```python
# Top three conssignment has total percentage count
df1['CumPct'] = np.cumsum(df1['pct'])
df1
```

## Descriptive Statistics
Central Tendency : Mean, Mode, Median

Dispersion or Variance or Speed : Range, Inter-Quartile Range, Variation and Standard Deviation

Shape : Skewness, Kurtosis

```python
# Mean of x*f
# Problem : Weighted Average/Mean
x = np.array([3,4,5,6])
f = np.array([20,18,9,12])

sum(x*f)/sum(f)
```

Mean:
- Mean Should be used when there are no extreme value in the data set
- Mean should be used when data is Symetrical
- Median can be used in all cases.

Mode: It can be used for Qualitative and it have multiple value as well.
* 1-Mode : Mode
* 2-Mode : Bi-Mode
* 3-Mode : Tri-Mode
* 3+ Mode : Multi-Mode
