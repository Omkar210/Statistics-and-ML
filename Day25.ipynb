# tanh function
The tanh (hyperbolic tangent) function is a foundational mathematical function that maps any real-valued input to a continuous output between -1 and 1
$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

## Derivative of tanh function
$$
\frac{d}{dx}\tanh(x) = 1 - \tanh^2(x)
$$

$$
-1 < \tanh(x) < 1
$$

# Huber Loss Function
- It provides a robust approach by combining the best properties of Mean Squared Error (MSE) and Mean Absolute Error (MAE).
- It is quadratic for small errors and linear for large errors, controlled by a threshold parameter δ.

## Condition:
$$
L_\delta(e)=
\begin{cases}
\frac{1}{n}\sum{(y_p - y_a)}^2, & error \le \delta \\
\delta\left(|{y_p - y_a}|-\frac{\delta}{2}\right), & error > \delta
\end{cases}
$$

# Momentum Optimizer
### Momentum Optimizer

**Momentum** is an improvement over standard **Gradient Descent** that helps the optimizer move faster toward the minimum and reduces oscillations.

---

## Problem with Gradient Descent

Standard Gradient Descent updates weights as:

```latex
W_{t+1} = W_t - \eta \nabla J(W_t)
```

where:

* (W_t) = current weights
* (\eta) = learning rate
* (\nabla J(W_t)) = gradient

Issue:

* Slow convergence
* Zig-zag movement in narrow valleys
* Can get stuck in local irregularities

---

## Idea Behind Momentum

Think of a ball rolling down a hill.

* The ball gains speed in directions where gradients consistently point the same way.
* Small fluctuations are smoothed out.
* Faster convergence toward the minimum.

Instead of using only the current gradient, Momentum also remembers past gradients.

---

## Mathematical Formula

Velocity update:

```latex
v_t = \beta v_{t-1} + (1-\beta)\nabla J(W_t)
```

Weight update:

```latex
W_{t+1} = W_t - \eta v_t
```

where:

* (v_t) = velocity
* (\beta) = momentum coefficient
* (\eta) = learning rate

Typically:

```latex
\beta = 0.9
```

or

```latex
\beta = 0.99
```

---

## Intuition

Suppose gradients over iterations are:

```text
2, 2, 2, 2, 2
```

Gradient Descent:

```text
step = 2 every time
```

Momentum:

```text
velocity accumulates

2
3.8
5.42
6.88
...
```

Updates become larger in the correct direction.

---

## Example

Assume:

```text
β = 0.9
η = 0.01
v₀ = 0
gradient = 5
```

Iteration 1:

```text
v₁ = 0.9(0) + 0.1(5)
   = 0.5
```

Weight update:

```text
W = W - 0.01 × 0.5
```

Iteration 2:

```text
v₂ = 0.9(0.5) + 0.1(5)
   = 0.95
```

Velocity keeps increasing when gradients point in the same direction.

---

## Advantages

### 1. Faster Convergence

Reaches minima faster than vanilla Gradient Descent.

### 2. Reduces Oscillations

Without momentum:

```text
/\/\/\/\/\
```

With momentum:

```text
--------
```

Smoother path toward the minimum.

### 3. Escapes Small Local Irregularities

Accumulated velocity can carry optimization through shallow local minima or noisy regions.

---

## Limitations

* Requires tuning ( \beta ).
* Can overshoot the minimum if learning rate is too high.
* Usually outperformed by modern adaptive optimizers such as:

  * RMSProp
  * Adam
  * AdamW

---

## Momentum vs Adam

| Feature                  | Momentum | Adam |
| ------------------------ | -------- | ---- |
| Uses past gradients      | ✅        | ✅    |
| Adaptive learning rate   | ❌        | ✅    |
| Faster than GD           | ✅        | ✅    |
| Most commonly used today | Less     | More |

---

## Interview Answer

**Momentum Optimizer** accelerates Gradient Descent by maintaining a velocity term that accumulates past gradients. This helps reduce oscillations, speeds up convergence, and allows the optimizer to move more efficiently toward the minimum. The update rule combines the current gradient with a fraction of the previous update, controlled by the momentum coefficient ( \beta ), typically set to 0.9.

# AdaGrad
# AdaGrad (Adaptive Gradient Algorithm)

**AdaGrad** is an optimizer that adapts the learning rate for each parameter individually.

The key idea is:

> Parameters that receive large gradients get smaller learning rates, while parameters with small gradients get larger learning rates.

This is particularly useful for **sparse data** such as text processing and NLP.

---

## Why AdaGrad?

Gradient Descent uses a fixed learning rate:

$$
W_{t+1} = W_t - \eta \nabla J(W_t)
$$

Problem:

* Same learning rate for all parameters.
* Difficult to choose the perfect learning rate.

AdaGrad automatically adjusts learning rates during training.

---

## Mathematical Formulation

### Accumulate Squared Gradients

$$
G_t = G_{t-1} + g_t^2
$$

where:

* $g_t$ = current gradient
* $G_t$ = accumulated squared gradients

### Weight Update

$$
W_{t+1} = W_t \frac{\eta}{\sqrt{G_t+\epsilon}} g_t
$$

where:

* $\eta$ = learning rate
* $\epsilon$ = small constant (e.g., (10^{-8}))
* $G_t$ = sum of past squared gradients

---

## How It Works

Suppose:

```text
Learning Rate = 0.1
```

Parameter A receives large gradients:

```text
5, 4, 6, 5
```

Squared gradient accumulation grows quickly:

```text
25 → 41 → 77 → 102
```

Effective learning rate becomes:

$$
\frac{0.1}{\sqrt{102}}
\approx 0.01
$$

Smaller updates.

---

Parameter B receives small gradients:

```text
0.1, 0.2, 0.1
```

Accumulation remains small.

Effective learning rate stays relatively large.

Result:

* Frequently updated parameters slow down.
* Rarely updated parameters continue learning.

---

## Advantages

### 1. Adaptive Learning Rate

Each parameter gets its own learning rate.

### 2. Good for Sparse Features

Works well in:

* NLP
* Recommendation systems
* Text classification

### 3. Less Manual Tuning

Requires less learning-rate tuning than vanilla Gradient Descent.

---

## Main Problem

The accumulated squared gradients keep growing:

$$
G_t = \sum_{i=1}^{t} g_i^2
$$

As training continues:

$$
\sqrt{G_t} \rightarrow \text{very large}
$$

Therefore:

$$
\frac{\eta}{\sqrt{G_t}} \rightarrow 0
$$

The learning rate becomes extremely small.

Eventually:

```text
Weight updates ≈ 0
```

Training almost stops.

This is AdaGrad's biggest weakness.

---

## Example

Assume:

```text
η = 0.1
Gradient = 2
```

Iteration 1:

$$
G_1 = 4
$$

$$
\text{LR}_{eff} = \frac{0.1}{\sqrt{4}} = 0.05
$$

Iteration 2:

$$
G_2 = 8
$$

$$
\text{LR}_{eff}
= \frac{0.1}{\sqrt{8}} = 0.035
$$

Iteration 10:

Learning rate becomes much smaller.

Training slows down significantly.

---

## AdaGrad vs Momentum

| Feature                               | Momentum | AdaGrad |
| ------------------------------------- | -------- | ------- |
| Uses past gradients                   | ✅        | ✅       |
| Adaptive learning rate                | ❌        | ✅       |
| Handles sparse data well              | ❌        | ✅       |
| Learning rate decreases automatically | ❌        | ✅       |
| Can become too small                  | ❌        | ✅       |

---

## AdaGrad vs RMSProp

AdaGrad accumulates **all past squared gradients**:

$$
G_t = G_{t-1} + g_t^2
$$

RMSProp uses an **exponentially weighted average**:

$$
G_t = \beta G_{t-1} + (1-\beta)g_t^2
$$

Because RMSProp "forgets" old gradients, its learning rate does not shrink to zero.

This largely fixes AdaGrad's main problem.

---

## Interview Answer

**AdaGrad (Adaptive Gradient Algorithm)** is an optimizer that adapts the learning rate for each parameter based on the history of squared gradients. Parameters with large gradients receive smaller updates, while parameters with small or infrequent gradients receive larger updates. It works particularly well for sparse datasets, but its learning rate continuously decreases over time, which can eventually make training very slow. This limitation led to improved optimizers such as RMSProp and Adam.

---

## RMSProp (Root Mean Square Propagation)

### Definition

RMSProp is an adaptive learning rate optimizer that improves AdaGrad by preventing the learning rate from shrinking too much over time.

Instead of storing all past squared gradients, RMSProp keeps an exponentially weighted moving average of squared gradients.

---

### Why RMSProp?

**Problem with AdaGrad:**

$$
G_t = G_{t-1} + g_t^2
$$

The accumulated squared gradients keep growing, causing the learning rate to become extremely small.

RMSProp solves this by using a moving average that gradually forgets old gradients.

---

### Step 1: Compute Moving Average of Squared Gradients

$$
S_t = \beta S_{t-1} + (1-\beta)g_t^2
$$

Rendered:

$$
S_t = \beta S_{t-1} + (1-\beta)g_t^2
$$

where:

* $S_t$ = moving average of squared gradients
* $g_t$ = current gradient
* $\beta$ = decay rate (typically 0.9)

---

### Step 2: Update Parameters

$$
W_{t+1} = W_t - \frac{\eta}{\sqrt{S_t+\epsilon}}g_t
$$

Rendered:

$$
W_{t+1} = W_t - \frac{\eta}{\sqrt{S_t+\epsilon}}g_t
$$

where:

* $W_t$ = current weight
* $\eta$ = learning rate
* $\epsilon$ = small constant (e.g., (10^{-8}))

---

### Complete RMSProp Algorithm

$$
S_t = 0.9S_{t-1} + 0.1g_t^2
$$

$$
W_{t+1} = W_t - \frac{\eta}{\sqrt{S_t+\epsilon}}g_t
$$


---

### Typical Hyperparameters

```text
Learning Rate (η) = 0.001
Decay Rate (β) = 0.9
ε = 1e-8
```

---

### Advantages

* Adaptive learning rate for each parameter
* Prevents learning rate from shrinking to zero
* Faster convergence than AdaGrad
* Works well for deep neural networks
* Handles noisy gradients effectively

---

### Disadvantages

* Requires tuning learning rate
* Does not use momentum directly
* Usually outperformed by Adam in many applications

---

### RMSProp vs AdaGrad

| Feature                       | AdaGrad | RMSProp |
| ----------------------------- | ------- | ------- |
| Adaptive learning rate        | ✅       | ✅       |
| Stores all past gradients     | ✅       | ❌       |
| Learning rate shrinks to zero | ✅       | ❌       |
| Suitable for deep learning    | Limited | ✅       |
| Faster convergence            | ❌       | ✅       |

---

### RMSProp vs Adam

| Feature                             | RMSProp | Adam |
| ----------------------------------- | ------- | ---- |
| Adaptive learning rate              | ✅       | ✅    |
| Momentum                            | ❌       | ✅    |
| Moving average of squared gradients | ✅       | ✅    |
| Most commonly used                  | Less    | More |

---

### Interview Answer

**RMSProp (Root Mean Square Propagation)** is an adaptive optimization algorithm that maintains an exponentially weighted moving average of squared gradients and uses it to adjust the learning rate for each parameter. Unlike AdaGrad, RMSProp prevents the learning rate from continuously decreasing, resulting in faster and more stable convergence during neural network training.

````python
df = pd.read_excel('CDAC_DataBook.xlsx', sheet_name='diabetes')
df.columns
````

````python
df = df[['Glucose','BloodPressure','Age','DietType','Outcome']]
diet_dummy = pd.get_dummies(df.DietType, drop_first=True).astype(int)
df = df.drop('DietType', axis=1)
df = pd.concat([df,diet_dummy], axis=1)
````

````python
x_train, x_test, y_train, y_test = train_test_split(df.drop('Outcome', axis=1), df.Outcome, test_size=0.2, random_state=20)
````

````python
classifier = Sequential()

classifier.add(Dense(units=5, activation='relu', input_dim=5))  # input layer of 5 neurons
classifier.add(Dense(units=7, activation='relu'))  # hidden layer of 7 neurons
classifier.add(Dense(units=5, activation='relu'))  # hidden layer of 5 neurons
classifier.add(Dense(units=1, activation='sigmoid'))  # output layer of 1 neuron
````

````python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

````python
mod1 = classifier.fit(x_train, y_train, batch_size=8,epochs=20)
````

````python
y_pred = classifier.predict(x_test)
y_pred = (y_pred >= 0.5)
````

````python
cm = confusion_matrix(y_test, y_pred)
print(cm)
````
