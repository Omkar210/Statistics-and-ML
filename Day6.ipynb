# Dispersion / Variance

### Range (DrawBacks):
- It can get affected by outliers
- It only consider 2 values from the list of thousands of values which is not recommended.
- Formula: Min - Max


### Inter-Quartile Range
- IQR = 75th %tile - 25th %tile

**%tile => i = 1 + (n-1)xp / 100**

**==> 25%tile = 1 + (23 - 1)x25 / 100**

75th %tile is 20 => 75% of values are less than 20

- Formula :
IQR = np.percentile(x1, 75) - np.percentile(x1,25)
